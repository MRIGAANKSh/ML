```python
import pandas as pd

# Creating the dataframe
data = {
    'Head Size(cm^3)': [1, 2, 3, 4, 5],
    'Brain Weight(grams)': [100, 200, 300, 400, 500]
}

df = pd.DataFrame(data)

# Collecting x and y
x = df['Head Size(cm^3)'].values
y = df['Brain Weight(grams)'].values

```

```python
#mean x y
mean_x=np.mean(x)
mean_y=np.mean(y)

#Total no. values
n=len(x)
#using formula to calculate b1 and b2
numr=0
denom=0
for i in range(n):
    numr+=(x[i] -mean_x)*(y[i]-mean_y)
    denom+=(x[i]-mean_x)**2
b1=numr/denom
b0=mean_y-(b1*mean_x)
print(b1,b0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming x and b0, b1 are defined
# Example:
# x = np.array([...])  # your x data points
# b0 = ...  # intercept
# b1 = ...  # slope

max_x = np.max(x) + 100
min_x = np.min(x) - 100

# Calculating line values for x and y
x_line = np.linspace(min_x, max_x, 1000)  # Creating x values for the line
y_line = b0 + b1 * x_line  # Calculate corresponding y values for the line

# Plotting the regression line
plt.plot(x_line, y_line, color='#58b970', label='Regression Line')

# Plotting the scatter plot (assuming y is the dependent variable)
plt.scatter(x, y, color='#ef5423', label='Scatterplot')

# Adding labels
plt.xlabel('Head Size in cmÂ³')
plt.ylabel('Brain Weight in grams')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

```python
#value of R2-how well model observe data ... for a perfect fit ..
ss_t=0
ss_r=0
for i in range(n):
    y_pred=b0+b1*x[i]
    ss_t+=(y[i]-mean_y)**2
    ss_r+=(y[i]-y_pred)**2
r2=1-(ss_r/ss_t)
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Reshaping x to be a 2D array (required for sklearn)
x = x.reshape(-1, 1)  # This assumes x is a 1D array and reshapes it to (n_samples, 1)

# Creating and fitting the model
reg = LinearRegression()
reg.fit(x, y)

# Making predictions
y_pred = reg.predict(x)

# Calculating R^2 score
r2 = r2_score(y, y_pred)
print(f"R^2: {r2}")
```
